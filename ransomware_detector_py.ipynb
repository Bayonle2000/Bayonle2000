```
import pandas as pd
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

# Create a dummy CSV file for demonstration purposes
dummy_data = {
    'packet_length': [100, 120, 150, 200, 110, 300, 80, 250, 130, 180],
    'inter_arrival_time': [0.1, 0.2, 0.15, 0.3, 0.25, 0.05, 0.4, 0.1, 0.22, 0.18],
    'entropy': [6.5, 6.8, 7.0, 7.2, 6.7, 5.5, 7.5, 6.0, 6.9, 7.1],
    'label': [0, 0, 0, 0, 0, 1, 0, 1, 0, 1] # 0 for Benign, 1 for Malicious
}
dummy_df = pd.DataFrame(dummy_data)
dummy_df.to_csv('network_logs.csv', index=False)

# 1. Load your processed PCAP data
# Features: packet_length, inter_arrival_time, protocol_type
data = pd.read_csv('network_logs.csv')

X = data[['packet_length', 'inter_arrival_time', 'entropy']]
y = data['label'] # 0 for Benign, 1 for Malicious

# 2. Train the Model
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# 3. Output Accuracy for Portfolio
print(classification_report(y_test, model.predict(X_test)))
```
